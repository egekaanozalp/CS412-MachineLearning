# The Context of the Analysis

In the following questions, while analyzing probabilities, the missing values of the datasets are ignored to avoid miscalculation. E.g.: Suppose that we are looking for the probability of people who are older than 30 and living in France. In this case, to avoid miscalculation (underestimation), I ignored the cases such as (nAn, France) or (35, nAn) since they increase the value of denominator while keeping the value of the numerator the same.

This issue can be handled by filling these values using the mean value or using kNN algorithm (or any other machine learning algorithms). However, such options are out of scope of this homework. Hence, these data points are ignored in the following calculations.

# Question 1

Question: What is the probability that a respondent is currently employed as a Programmer given they use C/C++ at work?

Mathematical Notation of the Question: P(Y='Programmer' | X='C/C++')

```python
# This cell focuses on X='C/C++':
c_cpp_df = multipleChoice_df[(multipleChoice_df["WorkToolsSelect"].
                             str.contains("C/C++", na=False, regex=False)) &
                              (multipleChoice_df["CurrentJobTitleSelect"].isnull()
                              ==False)][["CurrentJobTitleSelect", "WorkToolsSelect"]]
c_cpp_df
```

```python
# This cell focuses on Y='Programmer' | X='C/C++':
prog_df = c_cpp_df[c_cpp_df["CurrentJobTitleSelect"]=="Programmer"]
prog_df
```

```python
# This cell focuses on P(Y='Programmer' | X='C/C++'):
print("P(Y='Programmer' | X='C/C++') =", len(prog_df)/len(c_cpp_df))
```

# Question 3

Question: What is the probability that a respondent works in the Technology industry given that they earn more than 40,000 USD annually?

Mathematical Notation of the Question: P(Y="Technology" | X âˆˆ (40000, âˆž) USD)

Note: See that the value of 40,000 USD is not included since we are focusing on values more than 40,000 USD.

```python
# This function is created to convert the compensation amount values of the data
# points to USD. It is used with pd.apply function:
def convert_salary(row, exchange_rate_df):
    currency = row["CompensationCurrency"]
    exchange_rate_series = exchange_rate_df[exchange_rate_df["originCountry"] ==
                                            currency]["exchangeRate"]
    if not exchange_rate_series.empty:
        exchange_rate = float(exchange_rate_series.iloc[0])
        salary_usd = float(row["CompensationAmount"]) * exchange_rate
        row["CompensationAmount"] = salary_usd
        row["CompensationCurrency"] = "USD"
    else:
      print("Currency", currency, "is not in the list!")
    return row
```

```python
# This cell focuses on X âˆˆ (40000, âˆž) USD:
salary_df = multipleChoice_df[(multipleChoice_df["CompensationAmount"].isnull()==
                               False) & (multipleChoice_df["CompensationCurrency"].
                                         isnull()==False) & (multipleChoice_df
                                          ["EmployerIndustry"].isnull()==False)][
                                              ["EmployerIndustry",
                                               "CompensationAmount",
                                               "CompensationCurrency"]]

salary_df["CompensationAmount"] = pd.to_numeric(salary_df["CompensationAmount"],
                                                errors="coerce")

salary_df = salary_df.apply(lambda row: convert_salary(row, conversion_df), axis=1)
salary_df = salary_df[(salary_df["CompensationCurrency"] == "USD")]
salary_df = salary_df[salary_df["CompensationAmount"] > 40000]
salary_df = salary_df[salary_df["EmployerIndustry"].isnull()==False]
salary_df
```

```python
# This cell focuses on Y="Technology" | X âˆˆ (40000, âˆž) USD:
tech_df = salary_df[salary_df["EmployerIndustry"] == "Technology"]
tech_df
```

```python
# This cell focuses on P(Y="Technology" | X âˆˆ (40000, âˆž) USD):
print("P(Y='Technology' | X âˆˆ (40000, âˆž) USD) =", len(tech_df)/len(salary_df))
```

# Question 4

Question: What is the joint probability of a respondent being over 30 years old and having at least a Bachelorâ€™s degree?

Mathematical Notation of the Question: P(X_1 âˆ© X_2) where X_1 is "respondent being over 30 years old" and X_2 is "respondent having at least a Bachelorâ€™s degree".

```python
# See that the degrees that have more or equal importance to Bachelor's Degree
# form the set: {"Bachelor's degree", "Master's degree", "Doctoral degree",
# 'Professional degree'}
multipleChoice_df["FormalEducation"].unique()
```

```python
# This cell focuses on X_1 âˆ© X_2 which is respondents that are over 30 years old
# and possess at least a Bachelorâ€™s degree:
objective_degrees = ["Bachelor's degree", "Master's degree", "Doctoral degree",
                     "Professional degree"]
age_degree_df = multipleChoice_df[multipleChoice_df["FormalEducation"].
                                  isin(objective_degrees)][["FormalEducation",
                                                            "Age"]]
age_degree_df = age_degree_df[(age_degree_df["Age"]>30)]
age_degree_df
```

```python
# This cell focuses on the denominator of the probability function which is the
# number of data points which does not have missing value on columns 'Age' and
# 'FormalEducation':
denominator_df = multipleChoice_df[(multipleChoice_df["FormalEducation"].isnull()
                                   ==False) & (multipleChoice_df["Age"].isnull()
                                   ==False)]
denominator_df[["FormalEducation", "Age"]]
```

```python
# This cell focuses on P(X_1 âˆ© X_2):
print("P(X_1 âˆ© X_2) =", len(age_degree_df)/len(denominator_df))
```

# Question 6

Question: What is the joint probability that a respondent is from France, earns less than 100,000 USD annually, and uses Cross-Validation Often or Most of the time?

Question Notation: P(X_1 âˆ© X_2 âˆ© X_3) where X_1 is "respondent is from France", X_2 is "respondent earns less than 100,000 USD annually" and X_3 is "respondent uses Cross-Validation Often or Most of the time".

```python
# This cell focuses on X_1 âˆ© X_2 âˆ© X_3:
# Start with filtering X_2:
salary_df = multipleChoice_df[(multipleChoice_df["Country"].isnull()==False) &
                              (multipleChoice_df["CompensationAmount"].isnull()==False) &
                              (multipleChoice_df["CompensationCurrency"].isnull()==False) &
                              (multipleChoice_df["WorkMethodsFrequencyCross-Validation"].
                               isnull()==False)][
                                   ["Country", "CompensationAmount",
                                    "CompensationCurrency",
                                    "WorkMethodsFrequencyCross-Validation"]]

salary_df["CompensationAmount"] = pd.to_numeric(salary_df["CompensationAmount"],
                                                errors="coerce")

salary_df = salary_df.apply(lambda row: convert_salary(row, conversion_df), axis=1)
salary_df = salary_df[(salary_df["CompensationCurrency"] == "USD")]
salary_df = salary_df[salary_df["CompensationAmount"] < 100000]

# Continue with filtering X_1 and X_3:
salary_df = salary_df[(salary_df["Country"] == "France") &
                      (salary_df["WorkMethodsFrequencyCross-Validation"].
                       isin(["Often", "Most of the time"]))]
salary_df
```

```python
# This cell focuses on the denominator of the probability function which is the
# number of data points which does not have missing value on columns 'Country',
# 'CompensationAmount', 'CompensationCurrency' and 'WorkMethodsFrequencyCross-Validation':
denominator_df = multipleChoice_df[(multipleChoice_df["Country"].isnull()
                                   ==False) & (multipleChoice_df["CompensationAmount"].
                                               isnull()==False) & (multipleChoice_df
                                                ["CompensationCurrency"].isnull()==False) &
                                                (multipleChoice_df
                                                 ["WorkMethodsFrequencyCross-Validation"].
                                                  isnull()==False)][
                                                      ["Country", "CompensationAmount",
                                                       "CompensationCurrency",
                                                       "WorkMethodsFrequencyCross-Validation"]]

# One edge case is that SPL does not exist in the conversion_df. Hence we should
# ignore it (A total of 1 case):
denominator_df = denominator_df[(denominator_df["CompensationCurrency"]=="SPL")==False]

denominator_df
```

```python
# This cell focuses on P(X_1 âˆ© X_2 âˆ© X_3):
print("P(X_1 âˆ© X_2 âˆ© X_3) =", len(salary_df)/len(denominator_df))
```

# Question 8

Question: Given the probability of a respondent wearing glasses is 0.15, and the probability of a respondent wearing glasses given they have a PhD is 0.25, find the probability of a respondent having a PhD given that they wear glasses.

Mathematical Notation of the Question: P(X_1 | X_2) = P(X_2 | X_1) * P(X_1) / P(X_2) where X_1 is "respondent having a PhD" and X_2 is "respondent wears glasses".

What is given? P(X_2) = 0.15 and P(X_2 | X_1) = 0.25.

What is needed? P(X_1).

```python
# This cell focuses on X_1:
glasses_df = multipleChoice_df[multipleChoice_df["FormalEducation"]==
                               "Doctoral degree"]["FormalEducation"]
glasses_df
```

```python
# This cell focuses on the denominator of the probability function which is the
# number of data points which does not have missing value on column
# 'FormalEducation':
denominator_df = multipleChoice_df[multipleChoice_df["FormalEducation"].isnull()
                                   ==False]["FormalEducation"]
denominator_df
```

```python
# This cell focuses on P(X_1 | X_2) = P(X_2 | X_1) * P(X_1) / P(X_2):
P_X_1 = len(glasses_df)/len(denominator_df)
print("P(X_1) =", P_X_1)
print("P(X_1 | X_2) = P(X_2 | X_1) * P(X_1) / P(X_2) =", 0.25 * P_X_1 / 0.15)
```
